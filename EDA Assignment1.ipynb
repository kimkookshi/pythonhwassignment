```python
# How does the death rate vary with age?

import numpy as np
import pandas as pd

# Load the dataset
file_path ="heart_failure_clinical_records_dataset.csv"
df = pd.read_csv(file_path)

# Define age bins
age_bins = [30, 40, 50, 60, 70, 80, 90, 100]
age_labels = ['30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-100']

# Assign each patient to an age group
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Calculate death rate per age group
death_rate = df.groupby('age_group')['DEATH_EVENT'].mean().to_numpy()

# Display death rate for each age group
for age_group, rate in zip(age_labels, death_rate):
    print(f"Age Group {age_group}: Death Rate = {rate:.2f}")
```

```python
#que 4

import numpy as np
import pandas as pd

# Load the dataset
file_path ="heart_failure_clinical_records_dataset.csv"
df = pd.read_csv(file_path)

age_bins = [30, 40, 50, 60, 70, 80, 90, 100]
age_labels = ['30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-100']


df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

death_rate = df.groupby('age_group')['platelets'].mean().to_numpy()

for age_group, rate in zip(age_labels, death_rate):
    print(f"Age Group {age_group}:platelets = {rate:.2f}")
```

```python
#QUE 9



import pandas as pd
from scipy import stats  # Ensure this line is included


# Drop rows with missing values in relevant columns
df_clean = df.dropna(subset=['ejection_fraction', 'diabetes'])

# Separate ejection fractions based on diabetes status
ejection_fraction_with_diabetes = df_clean[df_clean['diabetes'] == 1]['ejection_fraction']
ejection_fraction_without_diabetes = df_clean[df_clean['diabetes'] == 0]['ejection_fraction']

# Perform t-test
t_stat, p_value = stats.ttest_ind(ejection_fraction_with_diabetes, ejection_fraction_without_diabetes)

# Print results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Significant difference in ejection fraction between patients with and without diabetes.")
else:
    print("No significant difference in ejection fraction between patients with and without diabetes.")



```

```python
#que 10


# Drop rows with missing values in relevant columns
df_clean = df.dropna(subset=['serum_creatinine', 'DEATH_EVENT'])

# Separate serum creatinine levels based on survival status
creatinine_survived = df_clean[df_clean['DEATH_EVENT'] == 0]['serum_creatinine']
creatinine_did_not_survive = df_clean[df_clean['DEATH_EVENT'] == 1]['serum_creatinine']

# Perform t-test
t_stat, p_value = stats.ttest_ind(creatinine_survived, creatinine_did_not_survive)

# Print results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Significant difference in serum creatinine levels between survivors and non-survivors.")
else:
    print("No significant difference in serum creatinine levels between survivors and non-survivors.")
```
